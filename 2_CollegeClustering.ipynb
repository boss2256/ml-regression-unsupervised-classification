## Step 1. Load data

```python
import pandas as pd

# Load the dataset
file_path = 'College_Data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

## Step 2: Exploratory Data Analysis (EDA)
Now, let's perform EDA to gain insights into the data. We'll start by creating two charts:

A histogram to observe the distribution of colleges by graduation rate.
A scatter plot comparing the student/faculty ratio and the percentage of alumni who donate, to explore potential relationships between these features.

```python
from sklearn.preprocessing import StandardScaler

# Set the first column as the index
data.set_index(data.columns[0], inplace=True)

# Encode 'Private' column (Yes=1, No=0)
data['Private'] = data['Private'].map({'Yes': 1, 'No': 0})

# Check for missing values
missing_values = data.isnull().sum()

# Normalize the numerical features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns, index=data.index)

# Display missing values information
missing_values, data_scaled.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress future warnings from seaborn and pandas
warnings.simplefilter(action='ignore', category=FutureWarning)

# Histogram for Graduation Rate
plt.figure(figsize=(7, 5))
sns.histplot(data['Grad.Rate'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Graduation Rates')
plt.xlabel('Graduation Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for Student/Faculty Ratio vs Percentage of Alumni who Donate
plt.figure(figsize=(7, 5))
sns.scatterplot(x='S.F.Ratio', y='perc.alumni', hue='Private', style='Private', data=data)
plt.title('Student/Faculty Ratio vs Alumni Donation Percentage')
plt.xlabel('Student/Faculty Ratio')
plt.ylabel('Percentage of Alumni who Donate')
plt.show()
```

* A histogram showing the distribution of graduation rates, with a kernel density estimate (KDE) overlay to show the distribution shape.* 
A scatter plot illustrating the relationship between the student/faculty ratio and the percentage of alumni who donate, with colleges differentiated by whether they are private or public.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Dropping non-numerical and identifier columns
data_for_clustering = data.drop(['Private'], axis=1)

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_for_clustering)

# Applying PCA to reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Converting to a DataFrame for easier plotting
data_pca_df = pd.DataFrame(data=data_pca, columns=['PC1', 'PC2'])

data_pca_df.head()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import numpy as np

# K-Means Clustering Plot
plt.figure(figsize=(7, 5))
sns.scatterplot(x='PC1', y='PC2', hue='KMeans', palette='viridis', data=data_pca_df)
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Agglomerative Clustering Plot
plt.figure(figsize=(7, 5))
sns.scatterplot(x='PC1', y='PC2', hue='Agglomerative', palette='viridis', data=data_pca_df)
plt.title('Agglomerative Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# DBSCAN Clustering Plot
plt.figure(figsize=(7, 5))
sns.scatterplot(x='PC1', y='PC2', hue='DBSCAN', palette='viridis', data=data_pca_df, legend="full")
plt.title('DBSCAN Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

1. K-Means Clustering: The plot shows how the data points are grouped into three clusters. K-Means has divided the data into relatively distinct groups based on their location in the reduced-dimensional space.

2. Agglomerative Clustering: Similar to K-Means, Agglomerative Clustering has also identified three clusters. The distribution might look slightly different due to the hierarchical nature of this method, which considers the proximity between data points in a more nuanced way.

3. DBSCAN Clustering: DBSCAN groups points based on density, marking as outliers those that are in low-density regions. The number of clusters and their shape can vary significantly from those formed by K-Means and Agglomerative Clustering, as seen in the plot. This method does not force every point into a cluster, which can be particularly useful for identifying outliers or for datasets where the cluster structure is not well-defined by centroids.

Each of these models offers a unique approach to clustering, highlighting different aspects of the data's structure. Depending on your specific goals and the characteristics of your data, one model may be more appropriate than the others.
